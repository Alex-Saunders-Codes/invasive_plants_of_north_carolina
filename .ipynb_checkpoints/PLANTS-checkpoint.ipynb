```python
def in_nc(string):
        if string.find('NC') == -1:
            return False
        else:
            return True
```

```python
def in_l48(string):
        if string.find('L48') == -1:
            return False
        else:
            return True
```

```python
df_native.head()
```

```python
df_native['In_NC'] = df_native['State_and_Province'].apply(in_nc)
df_native['Native_L48'] = df_native['Native_Status'].apply(in_l48)
df_native['Native_NC'] = df_native['Native_Status'].apply(in_nc)
```
